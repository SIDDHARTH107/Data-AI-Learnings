```python
# Check how many unique values are there in output column i.e 'quality' column
df['quality'].unique()
```

```python
# Check how many values belongs to each unique values
df['quality'].value_counts()
# Here 681 datapoints belongs to quality=5, 638 to 6, 199 to 7 and so on...
```

```python
# Now we can clearly see that this is an Imbalanced Dataset as they are not in proper proportion
# Check for missing values
df.isnull().sum()
```

```python
# So there are no missing values in the dataset
# Check for duplicate records
df.duplicated()
```

```python
# Here we have to remove True wala records cuz here True indicates the duplicated or repeated records
df[df.duplicated()]
```

```python
# So here these are all the duplicate records shown above (240 rows are duplicate records and we don't want to use them)
# Now we need to remove these records
df.drop_duplicates(inplace=True)
```
