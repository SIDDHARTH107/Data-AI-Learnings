## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

Synthetic data is information that's artificially manufactured rather than generated by real-world events. It's created algorithmically and is used as a stand-in for test data sets of production or operational data, to validate mathematical models and to train machine learning (ML) models.

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2,
                           n_clusters_per_class=2, n_redundant=0)
# Here I have specifically said to do 2 types of clusters
```

```python
X # Independent feature
```

```python
y # Dependent Feature
```

The problem we are solving here is a binary cassification.

```python
# Converting X feature into column dataframe
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
# hue will show us categories in scatterplot graph
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
# SVC = Support Vector CLassifier
```

```python
# Here, we are taking linear kernel specifically
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
svc.coef_
```

```python
## Prediction
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

Here, we are getting 88.8 accuracy

## Hyperparameter Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['linear']
              }
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
# verbose=3 means it will give full detail information of GridSearch CV technique
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```
