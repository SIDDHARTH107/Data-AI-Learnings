## Decision Tree Classifier Implementation With Post Preprunning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
# This dataset is used for better visualization of decision tree, like to which category which flower belongs.
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df
```

```python
dataset.target
```

```python
# Converting our dataset into Independent and dependent features
X=df.iloc[:,:-1] # All rows and columns except the last target column
y=dataset.target
```

```python
X,y
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
# No need to perform standardization in decision tree classfier
```

```python
## Post Pruning
treeclassifier=DecisionTreeClassifier() # Initializing
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
# By default it's using Gini Impurity Decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=False)
```

```python
# By default it's using Gini Impurity Decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Post Pruning with max_depth =2
treeclassifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## prediction
y_pred=treeclassifier.predict(X_test)
```

```python
y_pred
```

```python

from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

Accuracy is __98%__.
