If there are outliers present in the dataset, they may impact your analysis in several ways:

Skewness of distribution: Outliers can significantly skew the distribution of flight prices. This skewness may impact the assumptions of statistical models or analyses that assume normality or balanced data.

Influence on statistical measures: Outliers can affect summary statistics such as the mean and standard deviation. These measures may no longer accurately represent the central tendency or variability of the data.

Biased insights: Outliers can lead to biased insights or conclusions about the relationship between variables. They may have a disproportionate impact on the analysis, misleading interpretations, or affecting the validity of statistical tests.

Model performance: If you plan to build a predictive model using the dataset, outliers can impact the model's performance. Models may be sensitive to extreme values and may fail to generalize well to new data.

This code will output the number of missing values in each column of the dataset. If there are missing values, the count will be greater than zero.

The impact of missing values on your analysis depends on the specific variables affected and the extent of missingness. Here are a few ways missing values can impact your analysis:

Biased results: If missing values are not handled appropriately, they can introduce bias in your analysis. For example, if the ratings of popular app categories are missing, it could skew the analysis of app ratings by category.

Reduced sample size: Missing values reduce the number of available data points, which can reduce the statistical power and precision of your analysis. This reduction in sample size may limit the generalizability of your findings.

Incorrect conclusions: Missing values can lead to incorrect conclusions if they are not properly addressed. Ignoring missing values or using inappropriate techniques to handle them may introduce errors in your analysis and lead to inaccurate interpretations.

To mitigate the impact of missing values, you can consider different strategies, such as:

Data imputation: Replace missing values with estimated values based on other available information. This can be done using techniques like mean imputation, regression imputation, or advanced imputation methods like K-nearest neighbors.

Dropping missing values: If the proportion of missing values is relatively small and randomly distributed, you may choose to exclude rows with missing values from your analysis. However, this should be done cautiously to avoid bias.

Accounting for missingness: Instead of imputing or dropping missing values, you can treat missingness as a separate category or create indicator variables to account for missingness in your analysis.
