```python
# We can create classification related datasets by using make_classification libraries
from sklearn.datasets import make_classification
```

```python
# Generating a dataset
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
print(X)
print(y)
```

```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

Now, we can see it's a 2D dataset (two dimensional). 1 = green dots and 0 = blue dots

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  lr = 0.1 # Learning rate

  for i in range(100):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],weights))
    weights = weights + lr*(y[j]-y_hat)*X[j]

  return weights[0],weights[1:]
```

The above code is creating a function into which if we pass our data (X,y), it will return weights and intercept.


```python
np.insert(X,0,1,axis=1)
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```
